```python
import os
import csv
from bs4 import BeautifulSoup
import pandas as pd
import requests as rq
import time
import random
```

```python
movies = pd.DataFrame(pd.read_html("https://raw.githubusercontent.com/CriMenghini/ADM/master/2019/Homework_3/data/movies2.html")[0])
movies.drop('Id', inplace=True, axis = 1)
```

```python
for i in range(9906,len(movies)):
    try:
        response = rq.get(movies.URL[i])
    except rq.exceptions.RequestException as e:
        print(e)
        time.sleep(20*60 + 30)
        response = rq.get(movies.URL[i])
    soup = BeautifulSoup(response.text, 'html.parser')
    f = open('article_'+str(i)+'.html','w')
    f.write(str(soup.encode('ascii').decode('ascii')))
    f.close()
    time.sleep(random.choice(range(1,6)))
```
